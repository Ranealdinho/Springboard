```python
WEEKLY_BASE_URL = 'weekly/2019/week{}.csv'
YEARLY_BASE_URL = 'yearly/2019.csv'

def generate_df():
    df = pd.DataFrame()
    for week in range(1, 18):
        weekly_df = pd.read_csv(WEEKLY_BASE_URL.format(week))
        weekly_df['Week'] = week
        df = pd.concat([df, weekly_df])
    return df

df19 = generate_df()
```

```python
#sorting players by position, ranked by fantasy finish
print(df19.loc[df['Pos'] == 'RB'].groupby('Player')['HalfPPRFantasyPoints'].sum().sort_values(ascending=False).head(12))
print(df19.loc[df['Pos'] == 'WR'].groupby('Player')['HalfPPRFantasyPoints'].sum().sort_values(ascending=False).head(12))
```

```python
#Comparing positions
df2019 = df19.groupby('Pos')['HalfPPRFantasyPoints'].describe().reset_index()
df2019[(df2019['Pos'] == 'RB') | (df2019['Pos'] == 'WR') | (df2019['Pos'] == 'TE') | (df2019['Pos'] == 'QB')]
```

```python
#Group players by standard deviation of fantasy points
df19.groupby('Player')[['HalfPPRFantasyPoints']].std().sort_values(by='HalfPPRFantasyPoints', ascending=False).head(10)
```
