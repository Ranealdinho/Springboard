```python
#Sorting by danceability to understand this variable a bit more
df = df.sort_values('danceability', ascending = False)
df.head(10)
```

interesting to see danceability charted, the worst songs to dance to logically make sense as they are classical, or intro skits to albums with just words.

when charted in reverse, the songs with the highest danceability are rap or hip hop or EDM type songs. make sense. I will now explore energy with danceability

```python
##Outlier Analysis
q1 = df['energy'].quantile(0.25)
q3 = df['energy'].quantile(0.75)
iqr = q3-q1
lower= q1 - 1.5 * iqr
upper= q3 + 1.5 * iqr
plt.figure(figsize = (10,6))
sns.distplot(df['energy'],color='green',hist_kws=dict(edgecolor="black",linewidth=2))
plt.axvspan(xmin = lower,xmax= df['energy'].min(),alpha=0.2, color='red')
plt.axvspan(xmin = upper,xmax= df['energy'].max(),alpha=0.2, color='red')


#interesting to see that the energy for most of my music seems to be on the higher side
#most of the songs i listen to are for my gym playlists, which I like to have a lot of loud and energetic songs
```

```python
##Outlier Analysis
q1 = df['danceability'].quantile(0.25)
q3 = df['danceability'].quantile(0.75)
iqr = q3-q1
lower= q1 - 1.5 * iqr
upper= q3 + 1.5 * iqr
plt.figure(figsize = (10,6))
sns.distplot(df['danceability'],color='b',hist_kws=dict(edgecolor="black",linewidth=2))
plt.axvspan(xmin = lower,xmax= df['energy'].min(),alpha=0.2, color='red')
plt.axvspan(xmin = upper,xmax= df['energy'].max(),alpha=0.2, color='red')
```

```python
##Visualising Data
#Danceability and energy
pos_dance = df['danceability'] 
neg_dance = df['energy']
fig = plt.figure(figsize=(12,9))
plt.xlabel('Danceability')
plt.ylabel('Count')
plt.title('Song Danceability Distribution')
pos_dance.hist(alpha=0.5,bins=30,color='b',label='Danceability')
neg_dance.hist(alpha=0.5,bins=30,color='g',label='Energy')
plt.legend()
plt.show()
```

```python
x = df["danceability"].values
y = df["valence"].values

x = x.reshape(x.shape[0], 1)
y = y.reshape(y.shape[0], 1)

regr = linear_model.LinearRegression()
regr.fit(x, y)

fig = plt.figure(figsize=(6, 6))
fig.suptitle("Correlation between danceability and song mood")

ax = plt.subplot(1, 1, 1)
ax.scatter(x, y, alpha=0.5)
ax.plot(x, regr.predict(x), color="red", linewidth=3)
plt.xticks(())
plt.yticks(())

ax.xaxis.set_major_locator(ticker.MultipleLocator(0.1))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.02))

ax.yaxis.set_major_locator(ticker.MultipleLocator(0.1))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.02))

plt.xlabel("danceability")
plt.ylabel("valence")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,12))

sns.regplot(
    x= df['danceability'], 
    y= df['popularity'],
    color = 'Red',
    label = 'Danceability',
    scatter=True).set_title('Danceability, Acousticness vs Popularity', fontsize = 20)

sns.regplot(
    x= df['acousticness'], 
    y= df['popularity'],
    color = 'b',
    label = 'Acousticness',
    scatter=True)

ax.legend(fontsize = 20, loc = 'upper left')


#interesting to see Danceability and Acousticness have different effects when it comes to popularity
#the More danceable a song is the become more popular it will likely be, while songs with more acousticness are less popular
```

```python
fig, ax = plt.subplots(figsize=(20,12))

sns.regplot(
    x= df['speechiness'], 
    y= df['popularity'],
    color = 'g',
    label = 'Speechiness',
    scatter=True).set_title('Duration vs Popularity', fontsize = 20)

sns.regplot(
    x= df['instrumentalness'], 
    y= df['popularity'],
    color = 'b',
    label = 'Instrumentalness',
    scatter=True)

ax.legend(fontsize = 20, loc = 'upper left')

#when it comes to speechiness and instrumentalness, more of each seem to indicate a negative correlation,
#but only very slightly, could be a result of a large majority of songs with 0 instrumentalness
```
